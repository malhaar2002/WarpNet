```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# from google.colab.patches import cv2_imshow
```

```python
# Original Image
image1 = cv2.imread('WarpNet\images\original_image.png')
# resize image
image1 = cv2.resize(image1, (500, 500))

# Effects
effect1 = cv2.imread("images/old_image_textures/dots_1.jpg")
effect2 = cv2.imread("images/old_image_textures/folded_2.jpg")
effect3 = cv2.imread("images/old_image_textures/scratches_1.jpg")
effect4 = cv2.imread("images/old_image_textures/scribble_1.jpg")
effect5 = cv2.imread("images/old_image_textures/stain_2.jpg")
effect6 = cv2.imread("images/old_image_textures/dirt_2.jpg")


```

```python
# cv2_imshow(image1)
cv2.imshow("Original Image", image1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Resize all images
image1 = cv2.resize(image1, (500, 500))
effect1 = cv2.resize(effect1, (500, 500))
effect2 = cv2.resize(effect2, (500, 500))
effect3 = cv2.resize(effect3, (500, 500))
effect4 = cv2.resize(effect4, (500, 500))
effect5 = cv2.resize(effect5, (500, 500))
effect6 = cv2.resize(effect6, (500, 500))

effects = [effect1,effect2,effect3,effect4,effect5,effect6]
```

```python
# # Plotting all images in one row
# plt.figure(figsize=(18, 4))
# for i, img in enumerate(effects):
#     plt.subplot(1, len(effects), i+1)
#     plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
#     plt.axis('off')
# plt.show()



titles = ["Effect 1", "Effect 2", "Effect 3", "Effect 4", "Effect 5", "Effect 6"]


# Plotting all blended images in one row
plt.figure(figsize=(18, 4))
for i, img in enumerate(effects):
    plt.subplot(1, len(effects), i+1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(titles[i])
    plt.axis('off')

# plt.suptitle('Structured Noises', fontsize=16)
plt.subplots_adjust(wspace=0.05)  # Reduce gap between images

plt.show()
```

```python
# Blend all images
blended_images = []
ctr = 1
for i in effects:
    blended_image = cv2.addWeighted(image1, 0.5, i, 0.5, 0)
    blended_images.append(blended_image)
    cv2.imwrite(f"blended_image_{ctr}.jpg", blended_image)
    ctr += 1

```

```python
titles = ["Effect 1", "Effect 2", "Effect 3", "Effect 4", "Effect 5", "Effect 6"]


# Plotting all blended images in one row
plt.figure(figsize=(18, 4))
for i, img in enumerate(blended_images):
    plt.subplot(1, len(blended_images), i+1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(titles[i])
    plt.axis('off')

# plt.suptitle('Damaged Images', fontsize=16)
plt.subplots_adjust(wspace=0.03)  # Reduce gap between images

plt.show()
```
